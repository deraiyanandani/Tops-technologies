```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Car Rental Analysis

rental_days = np.array([
3,2,5,4,7,2,3,3,1,6,4,2,3,5,2,4,2,1,3,5,6,3,2,1,4,2,4,5,3,2,
7,2,3,4,5,1,6,2,4,3,5,3,2,4,2,6,3,2,4,5
])

mean_rental = np.mean(rental_days)
median_rental = np.median(rental_days)
mode_rental = stats.mode(rental_days, keepdims=True)

print("Car Rental Analysis")
print("Mean:", mean_rental)
print("Median:", median_rental)
print("Mode:", mode_rental.mode[0])
```

```python
# 2.Delivery Time Variability
delivery_times = np.array([
3,5,2,4,6,2,3,4,2,5,7,2,3,4,2,4,2,3,5,6,3,2,1,4,2,4,5,3,2,7,
2,3,4,5,1,6,2,4,3,5,3,2,4,2,6,3,2,4,5,3
])

range_delivery = np.ptp(delivery_times)
variance_delivery = np.var(delivery_times, ddof=1)
std_delivery = np.std(delivery_times, ddof=1)

print("\nDelivery Analysis")
print("Range:", range_delivery)
print("Variance:", variance_delivery)
print("Standard Deviation:", std_delivery)
```

```python
# 3. Monthly Revenue
revenue = np.array([120,150,110,135,125,140,130,155,115,145,135,130])

mean_revenue = np.mean(revenue)
range_revenue = np.ptp(revenue)

print("\nRevenue Analysis")
print("Average Revenue:", mean_revenue)
print("Range:", range_revenue)
```

```python
# 4. Fuel Efficiency Analysis
models = {
    "A": np.array([30,32,33,28,31,30,29,30,32,31]),
    "B": np.array([25,27,26,23,28,24,26,25,27,28]),
    "C": np.array([22,23,20,25,21,24,23,22,25,24]),
    "D": np.array([18,17,19,20,21,18,19,17,20,19]),
    "E": np.array([35,36,34,35,33,34,32,33,36,34])
}

print("\nFuel Efficiency Analysis")
for model, data in models.items():
    print(f"\nModel {model}")
    print("Mean:", np.mean(data))
    print("Range:", np.ptp(data))
    print("Variance:", np.var(data, ddof=1))
```

```python
# 5.Defect Analysis

defects = ['A','B','C','D','E','F','G']
frequencies = np.array([30,40,20,10,45,25,30])

plt.figure()
plt.bar(defects, frequencies)
plt.title("Defect Frequency Bar Chart")
plt.xlabel("Defect Type")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.hist(frequencies, bins=5)
plt.title("Defect Frequency Histogram")
plt.xlabel("Frequency")
plt.ylabel("Count")
plt.show()

print("Most Common Defect:", defects[np.argmax(frequencies)])
```

```python
# 6.Skewness & Kurtosis
ratings = np.array([
4,5,3,4,4,3,2,5,4,3,5,4,2,3,4,5,3,4,5,3,4,3,2,4,5,3,4,5,4,3,
3,4,5,2,3,4,4,3,5,4,3,4,5,4,2,3,4,5,3,4,5,4,3,4,5,3,4,5,4,3,
3,4,5,2,3,4,4,3,5,4,3,4,5,4,2,3,4,5,3,4,5,4,3,4,5,3,4,5,4,3,
3,4,5,2,3,4,4,3,5,4
])

print("\nSatisfaction Analysis")
print("Skewness:", stats.skew(ratings))
print("Kurtosis:", stats.kurtosis(ratings))
```

```python
# 7. Quartiles & Percentiles

weights = np.arange(55, 520, 5)

print("\nWeight Analysis")
print("Q1:", np.percentile(weights, 25))
print("Median:", np.percentile(weights, 50))
print("Q3:", np.percentile(weights, 75))

print("15th Percentile:", np.percentile(weights, 15))
print("50th Percentile:", np.percentile(weights, 50))
print("85th Percentile:", np.percentile(weights, 85))
```

```python
# 8. Correlation Coefficient
hours = np.array([10,12,15,18,20,22,25,28,30,32,35,38,40,42,45,48,50,52,55,58,60,62,65,68,70,72,75,78,80,82])
scores = np.array([60,65,70,75,80,82,85,88,90,92,93,95,96,97,98,99,100,102,105,106,107,108,110,112,114,115,116,118,120,122])

correlation = np.corrcoef(hours, scores)[0,1]
print("\nCorrelation:", correlation)
```

```python
# 9. Binomial Probability (At least 8 correct)
prob_binomial = sum(stats.binom.pmf(k, 10, 0.25) for k in range(8,11))
print("\nProbability (≥8 correct):", prob_binomial)
```

```python
# 10. Normal Distribution Probability
prob_normal = stats.norm.cdf(1100, 1000, 100) - stats.norm.cdf(900, 1000, 100)
print("\nProbability (900–1100 hours):", prob_normal)
```

```python
# 11. Confidence Interval (Mean Height)
mean = 170
std = 8
n = 100

ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=std/np.sqrt(n))
print("\n95% Confidence Interval:", ci)
```

```python
# 12. Hypothesis Test (One-sample t-test)
t_stat = (510 - 500) / (20/np.sqrt(25))
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=24))

print("\nHypothesis Test")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
# 13. Two-Sample Hypothesis (Template)
# Example template
group1 = np.random.normal(75, 10, 25)
group2 = np.random.normal(80, 10, 25)

t_stat, p_val = stats.ttest_ind(group1, group2)
print("\nTwo Sample Test p-value:", p_val)
```

```python
# 14. Confidence Interval (Proportion)
n = 500
x = 320
p_hat = x / n

ci_prop = stats.norm.interval(0.90, loc=p_hat, scale=np.sqrt(p_hat*(1-p_hat)/n))
print("\n90% Confidence Interval (Proportion):", ci_prop)
```

```python
# 15.Poisson Probability
prob_poisson = stats.poisson.pmf(3, 2)
print("\nPoisson Probability (exactly 3 defects):", prob_poisson)
```
